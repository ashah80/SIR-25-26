```python
import json
import pandas as pd
from pathlib import Path

TARGETS = Path("data/train_targets.json")
CSV = Path("data/all_real_with_split_fixed.csv")

if not TARGETS.exists() or not CSV.exists():
    raise FileNotFoundError("Missing inputs. Run previous notebooks first.")

df = pd.read_csv(CSV)
ref_map = df.groupby(['source_id','split'])['new_path'].apply(list).to_dict()

with open(TARGETS, "r") as f:
    targets = json.load(f)

missing = []
for t in targets:
    key = (t["id"], t["split"])
    if key not in ref_map or len(ref_map[key]) == 0:
        missing.append((t["file_path"], key))

print(f"{len(missing)} targets have no references in CSV (out of {len(targets)})")
if missing:
    print("Examples (file_path, (id,split)) - first 20:")
    for i, m in enumerate(missing[:20]):
        print(i, m)
```
